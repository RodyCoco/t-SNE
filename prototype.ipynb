```python
fig, ax = plt.subplots()
scatter = ax.scatter(tsne_digits[:,0], tsne_digits[:,1],
                    c=digit_class,
                     s=3,
                    cmap=cmap_light)
ax.legend(*scatter.legend_elements(),
            title="Classes",
            loc='lower center',
            ncol=5,
            fancybox=True,
            shadow=True,
            bbox_to_anchor=(0.5, -0.5)
            )
fig.tight_layout()
ax.axis('square')
plt.show()
```
